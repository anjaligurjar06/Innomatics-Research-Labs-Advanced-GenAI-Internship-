# Problem 1: Unique Words in a Sentence


```python
sentence = input("Enter sentence: ")
# Split the sentence into individual words
words = sentence.split()
# Convert list of words into set to remove duplicate words
unique_words = set(words)
# Print count of unique words
print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

# Problem 2: Highest Salary from Employee Data

```python
employees = {
    "Ravi": 75000,
    "Meena": 50000,
    "Anita": 68000,
    "Kiran": 72000
}

top_employee = ""
max_salary = 0
# Loop through dictionary to compare salaries
for person in employees:
    if employees[person] > max_salary:
        max_salary = employees[person]
        top_employee = person

print("Highest Salary:", top_employee, "-", max_salary)
```

# Problem 3: Find Maximum and Minimum Values

```python
values = [45, 22, 89, 10, 66]
maximum = values[0]
minimum = values[0]
# Loop through the list
for num in values:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num
print("List:", values)
print("Max:", maximum)
print("Min:", minimum)
```

# Problem 4: Count Products Above a Price Threshold

```python
items_price = [450, 1200, 899, 1500, 300]
count = 0
# Loop through each price
for price in items_price:
    if price > 1000:
        count = count + 1
print("Products above 1000:", count)
```

# Problem 5: Calculate Attendance Percentage

```python
attendance = ["P", "P", "A", "P", "P"]
present_count = 0
# Loop through attendance list
for day in attendance:
    if day == "P":
        present_count = present_count + 1
total_days = len(attendance)
# Calculate attendance percentage
percentage = (present_count / total_days) * 100
print("Attendance Percentage:", percentage)
```

# Problem 6: Remove Duplicate Phone Numbers

# Problem 7: Count Character Frequency

```python
word_input = "python"
result_dict = {}
# Loop through each character in string
for letter in word_input:
    if letter in result_dict:
        result_dict[letter] = result_dict[letter] + 1
    else:
        result_dict[letter] = 1
print(result_dict)
```
