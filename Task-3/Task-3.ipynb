# Problem Statement 1: Employee Performance Bonus Eligibility


```python
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}
# Step 1: Find the highest performance score
highest_score = max(employees.values())
top_performers = []
# Step 3: Loop through dictionary to check who has highest score
for name, score in employees.items():
   if score == highest_score:
        top_performers.append(name) 
print("Top Performers Eligible for Bonus:", 
      ", ".join(top_performers), 
      "(Score:", highest_score, ")")
```

# Problem Statement 2: Search Query Keyword Analysis

```python
query = "Buy mobile phone buy phone online"
query = query.lower()
for ch in ".,!?":
    query = query.replace(ch, "")
words = query.split()
freq = {}
# Count frequency of each word
for word in words:
    if word in freq:
        freq[word] += 1
    else:
        freq[word] = 1
result = {}
# Store only words appearing more than once
for word, count in freq.items():
    if count > 1:
        result[word] = count
print(result)
```

# Problem Statement 3: Sensor Data Validation

```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]
valid_readings = []
for i in range(len(sensor_readings)):
    if sensor_readings[i] % 2 == 0:
        valid_readings.append((i, sensor_readings[i]))
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

# Problem Statement 4: Email Domain Usage Analysis

```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

count = {}
for email in emails:
    domain = email.split("@")[1] 
    
    if domain in count:
        count[domain] += 1
    else:
        count[domain] = 1

total_users = len(emails)
for domain, count in count.items():
    percentage = (count / total_users) * 100
    print(f"{domain}: {int(percentage)}%")

```

# Problem Statement 5: Sales Spike Detection

```python
sales_list = [1200, 1500, 900, 2200, 1400, 3000]
total = 0
for i in sales_list:
    total += i
avg = total / len(sales_list)
# 30% above average
limit = avg * 1.3
for i in range(len(sales_list)):
    if sales_list[i] > limit:
        print("Day", i+1, ":", sales_list[i])
```

# Problem Statement 6: Duplicate User ID Detection

```python
users = ["user1", "user2", "user1", "user3", "user1", "user3"]

check = {}

# counting ids
for name in users:
    if name in check:
        check[name] = check[name] + 1
    else:
        check[name] = 1

# showing duplicates
for name in check:
    if check[name] > 1:
        print(name + " → " + str(check[name]) + " times")

```
