# 1. User Login Check

```python
username = "admin"
password = "1234"
#take username and password from user
entered_username=input("Enter Username")
entered_password=input("Enter Password")
#check if the entered username and password matches
if entered_username == username and entered_password == password:
    print("Login Successful")
else:
    print("Invalid Credentials")# If either username or password is wrong

```

# 2. Pass / Fail Analyzer

```python
marks = [45, 78, 90, 33, 60]
passed_students=0
failed_students=0
# Loop through each mark in the list
for i in marks:
    if i>=50:   ## If marks are greater than or equal to 50 â†’ Pass
        passed_students+=1
    else:
        failed_students+=1
## Printing final results
print("Total Students:", len(marks))
print("Total Passed Students:", passed_students)
print("Total Failed Students:", failed_students)
        
        
```

# 3. Simple Data Cleaner

```python
names = [" Alice ", "bob", " CHARLIE "]
# Empty list to store cleaned names
new_names = []
# Loop through each name in the list
for i in names:
    i = i.strip()  #Remove extra spaces from beginning and end using strip()
    
    i = i.lower()   # Convert the name to lowercase using lower()
    new_names.append(i)
print("Cleaned Names:", new_names)
```

# 4. Message Length Analyzer

```python
messages = ["Hi", "Welcome to the platform", "OK"]
for i in messages:
    length = len(i)
# Print the message and its length
    print("Message is:", i)
    print("Length of message is:", length)
# Check if message is longer than 10 characters
    if length > 10:
        print("Status: Message is longer than required (FLAGGED)")
    else:
        print("Status: Message length is acceptable")
    print("----------------------") 
    
  
```

# 5. Error Message Detector

```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
# Counter to store number of ERROR entries
errors = 0
# Loop through each log entry
for i in logs:
# Check if the log entry is equal to "ERROR"
    if i == "ERROR":
        errors += 1  
print("Total Errors :", errors)
```
