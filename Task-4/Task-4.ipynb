# Problem 1: Mobile Recharge Validation System


```python
def check_recharge(amount):
    plans = [199, 299, 399, 599]
    # minimum amount check
    if amount < 50:
        print("Recharge amount must be at least ₹50")
        return False
    if amount not in plans:
        print("Invalid plan selected")
        return False
    
    return True
# retry until valid recharge
while True:
    amt = int(input("Enter recharge amount: "))
    
    if check_recharge(amt):
        print("Recharge Successful ✅")
        break
    else:
        print("Please try again.\n")
```

# Problem 2: Inventory Reorder Alert System


```python
def check_stock(items):
    for product in items:
        if items[product] < 15:
            print(product + " → Reorder Alert")
        else:
            print(product + " → Stock OK")


store_data = {
    "Rice": 20,
    "Sugar": 10,
    "Oil": 25,
    "Salt": 8
}

check_stock(store_data)
```

# Problem 3: Student Result Processing System

```python
def check_result(marks_list):
    total = 0
    # loop to calculate total marks
    for m in marks_list:
        total += m
      # calculate average
    avg = total / len(marks_list)
    
    # check pass or fail
    if avg >= 50:
        return avg, "Pass"
    else:
        return avg, "Fail"


scores = [60, 45, 55, 70]

# calling function
average_marks, result = check_result(scores)

print("Average marks:", average_marks)
print(" Result:", result)
```

# Problem 4: Cab Fare Estimator with Retry Option


```python
def calculate_fare(distance, peak):
    base = 50
    rate = 12
    # basic fare calculation
    total_fare = base + (distance * rate)
    # add 25% extra if peak hour
    if peak == "yes":
        total_fare = total_fare + (0.25 * total_fare)
    
    return total_fare
while True:
    km = float(input("Enter distance in km: "))
    peak_time = input("Is it peak hour? (yes/no): ").lower()
    
    amount = calculate_fare(km, peak_time)
    
    print("Total Fare: ₹", amount)
    
    again = input("Do you want to calculate again? (yes/no): ").lower()
    
    if again != "yes":
        break
```

# Problem 5: Employee Attendance Eligibility

```python
def check_attendance(record):
    present_days = 0
    
    # count number of "P"
    for day in record:
        if day == "P":
            present_days += 1
    
    # calculate percentage
    percentage = (present_days / len(record)) * 100
    
    # check eligibility
    if percentage >= 75:
        return percentage, "Eligible"
    else:
        return percentage, "Not Eligible"

attendance_data = ["P", "A", "P", "P", "A", "P", "P", "P"]

percent, status = check_attendance(attendance_data)

print("Attendance Percentage:", percent)
print("Status:", status)
```

# Problem 6: Password Strength Checker


```python
def check_password(pwd):
    # check minimum length
    if len(pwd) < 8:
        return "Weak Password"
    
    has_digit = False
    has_special = False
    
    # loop through each character
    for ch in pwd:
        if ch.isdigit():
            has_digit = True
        if ch in "@#$":
            has_special = True
    
    # final check
    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"

user_input = input("Enter password: ")

result = check_password(user_input)

print(result)
```
